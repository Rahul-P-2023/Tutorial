```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Flight_Price.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df.isnull().sum()
```

```python
def newd(x):
    if x=="New Delhi":
        return 'Delhi'
    else:
        return x
df['Destination']=df['Destination'].apply(newd)
```

```python
df
```

```python
df["Journey_day"]=pd.to_datetime(df['Date_of_Journey'],format='%d/%m/%Y').dt.day
```

```python
df["Journey_month"]=pd.to_datetime(df['Date_of_Journey'],format='%d/%m/%Y').dt.month
```

```python
df
```

```python
df=df.drop('Date_of_Journey',axis=1)
```

```python
df
```

```python
df["Dep_hour"]=pd.to_datetime(df['Dep_Time']).dt.hour
```

```python
df["Dep_min"]=pd.to_datetime(df['Dep_Time']).dt.minute
```

```python
df
```

```python
df=df.drop('Dep_Time',axis=1)
```

```python
df
```

```python
df["Arrival_hour"]=pd.to_datetime(df['Arrival_Time']).dt.hour
```

```python
df
```

```python
df["Arrival_min"]=pd.to_datetime(df['Arrival_Time']).dt.minute
```

```python
df
```

```python
df=df.drop('Arrival_Time',axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
```

```python
li=['Airline','Source','Destination','Total_Stops']
```

```python
for i in li:
    j=0
    e=LE.fit_transform(df[i])
    df=df.drop(i,axis=1)
    df.insert(j,i,e)
    j+=1
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['Total_Stops','Destination','Source','Airline','Journey_day','Journey_month','Dep_hour','Dep_min','Arrival_hour','Arrival_min']]
```

```python
y=df[['Price']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=10, criterion='absolute_error',max_depth=3)
```

```python
rfr.fit(x_train,y_train)
```

```python
pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,pred)
```
